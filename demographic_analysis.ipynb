## Yaş Dağılımı

```python
gamesinfo['Age'].describe()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=gamesinfo, x='Age', bins=20, kde=True, color='#8DE5A1')
plt.title("Kullanıcı Yaş Dağılımı", fontsize=14)
plt.xlabel("Yaş")
plt.ylabel("Kullanıcı Sayısı")
plt.grid(True, linestyle="--", alpha=0.4)
plt.show()
```

```python
bins = [15, 25, 35, 49]
labels = ['15-25', '25-35', '35-49']
gamesinfo['Age_Group'] = pd.cut(gamesinfo['Age'], bins=bins, labels=labels, right=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))

# Kadınlar
plt.subplot(1, 2, 1)
sns.countplot(
    data=female_df,
    x='GameGenre',
    palette="pink"
)

plt.title("Kadınların Oyun Tercihleri")
plt.xlabel("Oyun Türü")
plt.ylabel("Kullanıcı Sayısı")
plt.xticks(rotation=45)
```

```python
plt.subplot(1, 2, 2)
sns.countplot(
    data=male_df,
    x='GameGenre',
    palette='Blues'
)
plt.title("👨 Erkeklerin Oyun Tercihleri")
plt.xlabel("Oyun Türü")
plt.ylabel("")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Oynama Süresi Dağılımları

```python
genre_avg = gamesinfo.groupby("GameGenre")["PlayTimeHours"].mean().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=genre_avg.values, y=genre_avg.index, palette="pastel")
plt.title("🎮 Oyun Türlerine Göre Ortalama Oynama Süresi")
plt.xlabel("Ortalama Süre (Saat)")
plt.ylabel("Oyun Türü")
plt.tight_layout()
plt.show()
```
